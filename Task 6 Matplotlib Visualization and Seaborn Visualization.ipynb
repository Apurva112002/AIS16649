## Matplotlib Visualization:

### Line Plot:

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('simple line plot')
plt.show()
```

### Bar plot:


```python
x=['A','B','C','D']
y=[10,20,15,25]
plt.bar(x,y)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

### Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
data=np.random.randn(100)
plt.hist(data,bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Histogram')
plt.show()
```

### Scatter Plot:

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.bar(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

### Pie Chart:


```python
sizes=[30,20,25,25]
labels=['A','B','C','D']
plt.pie(sizes,labels=labels)
plt.title('pe chart')
plt.show()
```

## Seaborn Visualization:


### Distribution plot:


```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
sns.histplot(tips['total_bill'],color='skyblue')
plt.xlabel('Bill')
plt.ylabel('Freq')
plt.title("Distribution plot")
plt.show()
```

### Pair Plot:

```python
import seaborn as sns
import matplotlib.pyplot as plt
tips=sns.load_dataset('tips')
sns.pairplot(tips)
plt.show()
```

### Count Plot:

```python
import seaborn as sns
data = sns.load_dataset('tips')
sns.countplot(x='day', data=data)
plt.title('Count Plot')
plt.show()
```

### Violin Plot:

```python
sns.violinplot(data=data, x='day', y='total_bill')
plt.title('Violin Plot')
plt.show()
```

### Box Plot:

```python
sns.boxplot(data=data, x='day', y='total_bill')
plt.title('Box Plot')
plt.show()
```
